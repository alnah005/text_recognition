```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = [10, 10]

from PIL import Image

from segmentation.imageModifiers import *
```

# Create example for segmentation

```python
def saveNP(arr, fn):
    im = Image.fromarray((arr * 255).astype(np.uint8)).convert("RGB")
    im.save(fn)
```

```python
# load image and convert to greyscale
example = Image.open("data/segment_example_pre.jpeg")
example = example.convert("L")
example.save("data/segment_example_grey.jpeg")
```

```python
# smooth image
example = np.array(example)
ex_smooth = smoothImg(example, 10)
saveNP(ex_smooth, "data/segment_example_smooth.jpeg")
```

```python
# project onto y axis, plot histogram, and save
proj = project(ex_smooth, "y")
matplotlib.rcParams['figure.figsize'] = [20, 10]
plt.plot(proj[::-1])
plt.show()
matplotlib.rcParams['figure.figsize'] = [10, 10]
```
