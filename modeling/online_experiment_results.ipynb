# Create and import data

** Load in data **

```python
# training csv with image locations and transcriptions
data = pd.read_pickle("tf_output/online_data/online_metrics1400.pkl")
```

# Show images

```python
def getrow(bunch, batch, pred="pred"):
    isbunch = data.bunch == bunch
    isbatch = data.batch == batch
    isepoch = data.epoch == 0
    ispred = data.pred == pred
    isbb = np.logical_and(isbunch, isbatch)
    isep = np.logical_and(isepoch, ispred)
    myloc = np.logical_and(isbb, isep)
    return data[myloc].iloc[0]
```

```python
# Choose which bunch to look at
bunch = 1300
# Create the dataset for that batch
create_ASM_batch(bunch, bunch+100, "../data")
```
